```python
import pytesseract
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
pytesseract.pytesseract.tesseract_cmd=r"C:\\Program Files\\Tesseract-OCR\\tesseract.exe"
```

```python
img=cv2.imread("D:\\buckaafiiciondo\\anthro.jpeg")
plt.imshow(img)
```

```python
imgtoconvert=pytesseract.image_to_string(img)
print(imgtoconvert)
```

```python
rect=pytesseract.image_to_boxes(img)
print(rect)
```

```python
imgH,imgW,_=img.shape
for boxes in rect.splitlines():
    boxes=boxes.split(' ')
    x,y,w,h=int(boxes[1]),int(boxes[2]),int(boxes[3]),int(boxes[4])
    cv2.rectangle(img,(x,imgH-y),(w,imgH-h),(0,255,0),5)
plt.imshow(img)
```

```python
cam=cv2.VideoCapture(0)
while True:
    ret,frame=cam.read()
    vidtoconvert=pytesseract.image_to_string(frame)
    vidrect=pytesseract.image_to_boxes(frame)
    imgH,imgW,_=frame.shape
    x1,y1,w1,h1=0,0,imgH,imgW
    for boxes in vidrect.splitlines():
        boxes=boxes.split(' ')
        x,y,w,h=int(boxes[1]),int(boxes[2]),int(boxes[3]),int(boxes[4])
        cv2.rectangle(frame,(x,imgH-y),(w,imgH-h),(0,255,0),5)
    cv2.putText(frame,vidtoconvert,(x1 + int(w1/50),y1 + int(h1/50)),cv2.FONT_HERSHEY_SIMPLEX,0.7,(0,255,0),2)
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.imshow("Text Detection",frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```
